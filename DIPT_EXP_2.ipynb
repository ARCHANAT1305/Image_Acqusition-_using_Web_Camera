```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from google.colab import files
uploaded = files.upload()
filename = list(uploaded.keys())[0]
img = cv2.imread(filename)
cv2.imwrite("output_frame.jpg", img)
print("Image saved as output_frame.jpg")
print("Original Image:")
cv2_imshow(img)
resized = cv2.resize(img, (200, 200))
top_row = np.hstack((resized, resized))
bottom_row = np.hstack((resized, resized))
grid_resized = np.vstack((top_row, bottom_row))
print("Resized Grid:")
cv2_imshow(grid_resized)
rotated = cv2.rotate(img, cv2.ROTATE_180)
rotated_resized = cv2.resize(rotated, (200, 200))
top_row_rot = np.hstack((rotated_resized, rotated_resized))
bottom_row_rot = np.hstack((rotated_resized, rotated_resized))
grid_rotated = np.vstack((top_row_rot, bottom_row_rot))
print("Rotated Grid:")
cv2_imshow(grid_rotated)
```
